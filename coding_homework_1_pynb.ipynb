```python
# In this cell I used pandas to simplify the two sets of imported raw data that was in csv format
url_1 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
pandas_data1 = pd.read_csv(url_1)

url_2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
pandas_data2 = pd.read_csv(url_2)
```

```python
# In this cell I used the provided syntax to convert the two sets of pandas data into lists
data1 = pandas_data1.values.tolist()
print(data1)

data2 = pandas_data2.values.tolist()
print(data2)
```

```python
# In this cell I used the squeeze method in order to tidy up the array of the now listed data by getting rid of the brackets around each individual element
data1 = np.squeeze(data1)
print(data1)
```

```python
# Like the cell above for data1, I used the squeeze method in order to tidy up the array of the now listed data by getting rid of the brackets around each individual element
data2 = np.squeeze(data2)
print(data2)
```

```python
# I used the length function in order to display the length of data1
print(len(data1))
```

```python
# I used slicing syntax in order to display the 3rd, 4th, 5th and 6th element in data2
selected_elements = data2[2:6]
print(selected_elements)
```

```python
# I used negative indexing, [-2], [-1], in order to display the second-to-last element and last element in data1
secondto_last = data1[-2]
print(secondto_last)

last_element = data1[-1]
print(last_element)
```

```python
# I used the type function in order to display the type of data data2 is
print(type(data2))
print(data2.dtype)
```

```python
# I used the syntax given to create a histogram of data1
plt.hist(data1)
```

```python
# I recreacted the data1 histogram but included the xlabel and ylabel functions to add labels on the axes
# x-axis is resting membrane potential, how many times that resting membrane potential appears in the data set
plt.hist(data1)
plt.xlabel('Resting Membrane Potential (mV)')
plt.ylabel('Count')
```

```python
# As I did for data1, I used the syntax provided to create a histogram of data2
plt.hist(data2)
```

```python
# As I did for data1, I used the xlabel and ylabel functions to add labels on the axes
plt.hist(data2)
plt.xlabel('Resting Membrane Potential (mV)')
plt.ylabel('Count')
```

In the histogram displying data2, the cells appear to have an overall lower, or more polarized, resting membrane potential than the data1 cells. In the data2 histogram, the data is shifted more towards the left, displaying more negative resting mebrane potentials in comparison to data1.
