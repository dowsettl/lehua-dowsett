```python
# I used the import function to import numpy, matplotlib.pyplot and scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# I used numpy in order to simplify the two sets of raw data that was in csv format into arrays.
url_1 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest_sst = np.loadtxt(url_1)

url_2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest_pvalb = np.loadtxt(url_2)
```

```python
# I used np.numpy syntax to find the mean of both v_rest_sst and v_rest_pvalb.
mean_sst = np.mean(v_rest_sst)
print(mean_sst)
mean_pvalb = np.mean(v_rest_pvalb)
print(mean_pvalb)

# I used the stats.ttest_in function to perform a t-test to establish if the means from v_rest_sst and v_rest_pvalb are statistically significant and come from different distributions.
t_stats,p_val = stats.ttest_ind(v_rest_sst,v_rest_pvalb)
print(p_val)

# I used and if statement and the provided fig, ax = plt.subplots(1,1) function in order to create two histograms for each set of data if the p-value of the t-test was < 0.05.
if p_val < 0.05:
  fig, ax = plt.subplots(1,2)
  # plots sst
  ax[0].hist(v_rest_sst, color='blue', edgecolor='black')
  ax[0].set_xlabel('Resting Potential (mV)')
  ax[0].set_ylabel('Frequency')
  ax[0].set_title('v_rest_sst')
  # plots pvalb
  ax[1].hist(v_rest_pvalb, color='yellow', edgecolor='black')
  ax[1].set_xlabel('Resting Potential (mV)')
  ax[1].set_ylabel('Frequency')
  ax[1].set_title('v_rest_pvalb')

  plt.show()





```

```python
# Using the def function, I defined the vrestHist() to execute the code from above.
def vrestHist():
    """
    Performs a t-test on v_rest_sst and v_rest_pvalb and
    plots histograms if the p-value is significant.
    """
    # finds the means
    mean_sst = np.mean(v_rest_sst)
    mean_pvalb = np.mean(v_rest_pvalb)

    # performs the t-test
    t_stat, p_val = stats.ttest_ind(v_rest_sst, v_rest_pvalb)

    # I used the if statement to make two subplots, one for v_rest_sst and one for v_rest_pvalb
    if p_val < 0.05:
        fig, ax = plt.subplots(1,2)

        # plots sst
        ax[0].hist(v_rest_sst, color='blue', edgecolor='black')
        ax[0].set_title('v_rest_sst')
        ax[0].set_xlabel('Resting Potential (mV)')
        ax[0].set_ylabel('Frequency')

        # plots pvalb
        ax[1].hist(v_rest_pvalb, color='yellow', edgecolor='black')
        ax[1].set_title('v_rest_pvalb')
        ax[1].set_xlabel('Resting Potential (mV)')
        ax[1].set_ylabel('Frequency')

        plt.show()
```

```python
# In this code cell I called the vrestHist() function that I just defined.
vrestHist()
```

```python
# Using the def function, I defined the vrestScatter() to execute the code from above.
def vrestScatter():
    """
    Creates a scatter plot of v_rest_sst and v_rest_pvalb and
    displays certain dot colors depending on the resting membrane potential voltage.
    """
    s_p_len = len(v_rest_sst)

    for i in range(s_p_len):
      x = v_rest_sst[i]
      y = v_rest_pvalb[i]

      if x > -60 and y > -60:
        dot_color = 'red'
      elif x > -60 and y < -60:
        dot_color = 'green'
      elif x < -60 and y < -60:
        dot_color = 'blue'
      elif x < -60 and y > -60:
        dot_color = 'purple'

      plt.scatter(x, y, color = dot_color)

    plt.xlabel('v_rest_sst (mV)')
    plt.ylabel('v_rest_pvalb (mV)')

    plt.show()
```
