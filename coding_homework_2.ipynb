```python
# I imported numpy, matplotlib and statistics to use for the imported data.
import numpy as np
from matplotlib import pyplot as plt
import statistics as st
```

```python
# I used numpy in order to simplify the two sets of raw data that was in csv format into arrays.
url_1 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest1 = np.loadtxt(url_1)

url_2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest2 = np.loadtxt(url_2)
```

```python
# I used the np.array() function in order to concatenate the two sets of data into a matrix with two rows and 500 columns. I then double checked the shape using the .shape syntax.
v_rest = np.array([v_rest1,v_rest2])
print(v_rest)
v_rest.shape
```

```python
# I used the .size method to determine the size of the, now combined, data.
v_rest.size
```

```python
# I used indexing in order to display the first row of the v_rest data. I used 0 because indexing starts at 0.
first_row = v_rest[0]
print(first_row)
```

```python
# I used the provided sum syntax with indexing to sum together the third column of the data. The [start:stop] was used to get both rows and then used the comma to get the third column to complete the index.
v_rest[0:2,2].sum()
```

```python
# I used indexing to get the second to last element (-2) from the second row (1).
secondto_last2 = v_rest[1,-2]
print(secondto_last2)
```

```python
# I used the syntax given to create a histogram of v_rest. I then used the xlabel and ylabel functions to add labels on the axes.
plt.hist(v_rest[0,:])
plt.xlabel('Resting Membrane Potential (mV)')
plt.ylabel('Count')
```

```python
# I used the provided syntax in order to obtain the mean of the values in the first row and then the mean values for the second row. The : indicates all columns.
v_rest_mean_sst = st.mean(v_rest[0,:])
print(v_rest_mean_sst)

v_rest_mean_pvalb = st.mean(v_rest[1,:])
print(v_rest_mean_pvalb)
```

The mean resting memebrane potential of row 2, or pvalb-containing neurons, is more negative than that of row 1, or sst-containing neurons.

```python
# The scipy package was imported in order to perform a t-test on the v_rest data.
import scipy.stats as stats
```

```python
# I used indexing in order to separate the two rows of data which represent the two different marker genes. Then I performed a t-test using the provided t-test function.
row_1 = v_rest[0,:]

row_2 = v_rest[1,:]

stats.ttest_ind(row_1, row_2)
```

After completing the t-test, the resulting p-value was approximately 7.1598e-09. Based on this very small p-value, there is a high likelihood that the difference in resting membrane potentials of sst and pvalb is statistically significant because 7.1598e-09 is much smaller than 0.05. This indicates that is very unlikely that the difference between restinging membrane potentials of sst-containing neurons and pvalb-containing neurons happened by random chance.
